***APLICACIÓN DE PROBABILIDAD***

Se Realiza el lanzamiento de un dado EQUILIBRADO JUSTO, de manera reiterada con las mismas condiciones, y se anotan sus resultados= CARA O CRUZ.

Se le pide al usuario que defina la cantidad N de lanzamientos y se calcula su frecuencia relativa= 'PROBABILIDAD' de obtener cara o cruz. Se obtiene los siguientes resultados:

```python
import numpy as np
n=int(input('ingrese la cantidad de simulación o lanzamientos'))

#========= FORMA 1 DE SIMULAR LANZAMIENTO DE UN DADO============
dado=[]
cara=0
cruz=0
for i in range(n):
  dado.append(np.random.rand(1))
  dado1=np.array(dado)
  if dado1[i]<0.5:
    #print('salio cara')
    #print(dado[i])
    cara +=1
  else:
    #print('salio cruz')
    #print(dado[i])
    cruz +=1
```

```python
#  mostrar la probailidad frecuencista de el lanzamiento de un dado
prom_cara=cara/n
prom_cruz=cruz/n
print('la probabilidad aproximada en '+str(n)+' lanzamientos de sacar CARA es:', prom_cara)
print('la probabilidad aproximada en '+str(n)+' lanzamientos de sacar CRUZ es:', prom_cruz)
```

```python
#========== FORMA COMPACTA DE SIMULAR USANDO NUMPY===========
m=int(input('ingrese la cantidad de simulación o lanzamientos'))
lanzamientos = np.random.randint(0, 2, m)

# Calcular las probabilidades

probabilidad_cara = np.sum(lanzamientos ==0 ) / len(lanzamientos)
probabilidad_sello = np.sum(lanzamientos== 1) / len(lanzamientos)

print(f"Probabilidad de cara: {probabilidad_cara:.2f}")
print(f"Probabilidad de sello: {probabilidad_sello:.2f}")
```

```python
import matplotlib.pyplot as plt
x=np.array(range(2))
y=np.array([probabilidad_cara, probabilidad_sello])

plt.bar(x,y)
plt.xlabel('Opción')
plt.ylabel('Frecuencia relativa')
plt.title('Histograma del lanzamiento de una moneda equilibrada')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ventas diarias de 4 productos durante una semana

ventas_diarias = np.array(
    #Lun,Mar,Mie,Jue,Vie,Sab,Dom
    [[20, 15, 25, 30, 18, 22, 24],  #Producto A
    [12, 20, 14, 8, 15, 18, 16],    #Producto B
    [35, 28, 32, 30, 26, 24, 30],   #Producto C
    [40, 38, 45, 42, 39, 41, 37]    #Producto D
    ]
)

print("Matriz de ventas diarias:",ventas_diarias)

productos = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
dias_semana = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

# Calculamos la media en las ventas por cada producto
media_productos = np.mean(ventas_diarias, axis=1)

print("Media ventas cada producto ")

for i, producto in enumerate(productos):
    print(f"{producto}: {media_productos[i]}")

# Calcular la mediana en las ventas por cada producto
mediana_productos = np.median(ventas_diarias, axis=1)


print("Mediana ventas cada producto ")

for i, producto in enumerate(productos):
    print(f"{producto}: {mediana_productos[i]}")


#graficamos (barra) la media y la mediana de las ventas de cada producto


#plt.figure(8,6)
plt.bar(productos,media_productos,edgecolor='black', color='blue')
plt.title('Media de ventas por cada producto')
plt.xlabel('Productos')
plt.ylabel('Ventas por producto')
plt.grid(True)
plt.show()

plt.bar(productos,mediana_productos,edgecolor='black', color='skyblue')
plt.title('Media y mediana de ventas por cada producto')
plt.xlabel('Productos')
plt.ylabel('Ventas por producto')
plt.grid(True)
plt.show()


# Calculamos la media en las ventas por dia
media_ventas_dia = np.mean(ventas_diarias, axis=0)


print("Media de ventas por dia")

for i, dia in enumerate(dias_semana):
    print(f"{dia}: {media_ventas_dia[i]}")

# Calculamos la mediana en las ventas por dia
mediana_ventas_dia = np.median(ventas_diarias, axis=0)


print("Mediana de ventas por dia")

for i, dia in enumerate(dias_semana):
    print(f"{dia}: {mediana_ventas_dia[i]}")

#Graficamos (trazo de lineas)  la media y mediana de ventas por dia

plt.plot(dias_semana,media_ventas_dia)
plt.title('Media de ventas por dia de la semana')
plt.xlabel('Dias de la semana')
plt.ylabel('Ventas diarias')
plt.grid(True)
plt.show()


plt.plot(dias_semana,mediana_ventas_dia,color="orange")
plt.title('Mediana de ventas por dia de la semana')
plt.xlabel('Dias de la semana')
plt.ylabel('Ventas diarias')
plt.grid(True)
plt.show()


#Conclusiones y sugerencias de estrategias o inversiones para mejorar el negocio

'''
Del análisis de las graficas de media y mediana de ventas de cada producto
y graficas de la media y mediana de ventas por dia  podemos observar que
en promedio el producto menos vendido es el producto B, luego sigue el A , el C y el D en orden de ventas.
Notamos también que el volumen de ventas de productos en general es menor durantes el dia viernes
y en segundo lugar los dias martes, observando un pico de ventas los dias miercoles (su media) y jueves (mediana)
el volumen de ventas va aumentando el  dia sabado y más aun el domingo.

De acuerdo a lo observado ,planteamos como estrategias de venta o inversion para potenciar el negocio:

-Aumentar la inversión en publicidad y marketing para los productos con menor venta ejemplo el producto B,
a fin de mejorar las ventas, y tambien  para el producto A.

-Implementación de ofertas y promociones con descuentos de precio de los productos en general  durante los dias de
menor venta, esto es los dias viernes y alternativamente los dias martes.

-Realizar una mejor gestion del stock de productos, a fin de disponer de suficientes unidades para la venta,
focalizando en los productos de mayor venta.

-Evaluar la posibilidad de ofrecer descuentos extra para estimular la salida de los productos de menor venta (ejemplo producto B)

-Ofrecimiento de packs de productos (producto de mayor venta junto a producto de menor venta) a precio promocional.

-También se deberia evaluar (deberíamos contar con el dato de precios unitarios de cada producto) las estrategias o
inversiones más convenientes de acuerdo al precio de cada producto,teniendo en cuenta que procentaje
representa  cada uno ellos en la ganancia de la empresa, si la venta de un producto en particular estimula la venta de otro,etc.

'''

# Calculamos el desvío estándar de los datos por cada producto
desvio_estandar_productos = np.std(ventas_diarias, axis=1)

indice_mayor_desvio = np.argmax(desvio_estandar_productos)
indice_menor_desvio = np.argmin(desvio_estandar_productos)


# Imprimir resultados de desvio estandar por producto
print("Desvío estándar por producto:")
for i, producto in enumerate(productos):
    print(f"{producto}: {desvio_estandar_productos[i]}")

print("\nProducto con mayor desvío estándar:", productos[indice_mayor_desvio])
print("Producto con menor desvío estándar:", productos[indice_menor_desvio])


#Graficamos datos de ventas de la semana de los productos con mayor y menor desvio estandar


plt.plot(dias_semana,ventas_diarias[indice_mayor_desvio],label=productos[indice_mayor_desvio])
plt.plot(dias_semana,ventas_diarias[indice_menor_desvio],label=productos[indice_menor_desvio])

plt.title('Ventas semana producto con mayor y menor desvio estandar')
plt.xlabel('Dias de la semana')
plt.ylabel('Ventas diarias')
plt.legend()
plt.grid(True)
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook

def creacion_matriz(cantidad):
  #generamos los datos random del sexo
  sexo=np.random.choice(["MASCULINO","FEMENINO","NO ESPECIFICA"],cantidad,[48.22,51.76,0.02])

  #generamos datos random de edad
  edad=np.random.normal(44,6,cantidad)
  edad=np.clip(np.around(edad),28,60)

  #generamos datos random de altura
  altura=np.random.normal(180,6,cantidad)
  altura=np.clip(np.around(altura,decimals=1),165,210)

  #generamos datos random de peso
  peso=np.random.normal(80,10,cantidad)
  peso=np.clip(np.around(peso, decimals=2),50,110)

  #generamos datos random de horas de descanso
  horas_descanso=np.random.normal(4,1.5,cantidad)
  horas_descanso=np.clip(np.around(horas_descanso, decimals=2),1,8)


  #generamos datos random de horas de trabajo
  horas_trabajo=np.random.normal(4,2,cantidad)
  horas_trabajo=np.clip(np.around(horas_trabajo, decimals=2),2,12)


  #generamos datos random de horas de ejercicio
  horas_ejercicio=np.random.normal(2,0.5,cantidad)
  horas_ejercicio=np.clip(np.around(horas_ejercicio, decimals=2),0.5,3)

  horas_descanso=24-horas_trabajo-horas_ejercicio
  return np.column_stack((sexo,edad,altura,peso,horas_descanso,horas_trabajo,horas_ejercicio))


cant_datos= int(input('Ingrese la cantidad de datos a cargar: '))
#llamamos matriz
poblacion=creacion_matriz(cant_datos)

#sacamos el genero del array
poblacion_sin_genero=poblacion[:,1:].astype(float)

#guardamos los array de peso y altura en sus respectivas variables
altura=poblacion_sin_genero[:,1]
peso=poblacion_sin_genero[:,2]

#calculamos la media de los datos
media_total=np.mean(poblacion_sin_genero, axis=0)

#imprimimos la media de los datos
print("La media de altura es : ",round((media_total[1]), 2),"La media de peso es : ",round((media_total[2]), 2),"La media de horas de descanso es : ",round((media_total[3]), 2),"La media de horas de trabajo es : ",round((media_total[4]), 2),"La media de horas de ejercicio es : ",round((media_total[5]), 2))

#calculamos la variabilidad
var_altura = np.std(altura)
print('Variabilidad altura ',var_altura)
var_peso = np.std(peso)
print('Variabilidad peso ',var_peso)

#creamos el histograma de altura
plt.figure(figsize=(6,4))
plt.hist(altura, color='cyan', ec='black',bins=20)
plt.axvline(media_total[1], color='red', linestyle='dashed', linewidth=2, label='Media altura')
plt.ylabel("Personas")
plt.xlabel("Altura")
plt.title('Altura de la Poblacion')
plt.grid(True)
plt.show()

#generamos el histograma de peso
plt.figure(figsize=(6,4))
plt.hist(peso, color='lightgreen', ec='black', bins=20)
plt.axvline(media_total[2], color='red', linestyle='dashed', linewidth=2, label='Media peso')
plt.ylabel("Personas")
plt.xlabel("Peso")
plt.title('Peso de la Poblacion')
plt.grid(True)
plt.show()

# Datos
x = peso
y = altura

# Gráfico
fig, ax = plt.subplots()
ax.scatter(x = x, y = y,cmap='viridis')
ax.set_xlabel('Peso')
ax.set_ylabel('Altura')
# plt.show()

# Figura 3D trabajo/descanso y ejercicio
horas_descanso = poblacion_sin_genero[:,3]
horas_trabajo_estudio = poblacion_sin_genero[:,4]
horas_ejercicio = poblacion_sin_genero[:,5]

# Crear una figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


# Crear el gráfico 3D
ax.scatter(horas_descanso, horas_trabajo_estudio, horas_ejercicio, cmap='viridis', s=100)

# Etiquetas de los ejes
ax.set_xlabel('Horas Descanso')
ax.set_ylabel('Horas Trabajo/Estudio')
ax.set_zlabel('Horas Ejercicio')
ax.set_title('Gráfico 3D de Distribución de Horas')

# Mostrar el gráfico
ax.grid
plt.grid(True)
plt.show()
```

Se realiza una encuesta a N usuario donde se registran los siguientes datos.

['DNI','GASTOS MENSUALES DE HOGAR','INGRESOS MENSUALES POR HOGAR', 'CANTIDAD DE MIEMBROS QUE VIVEN POR HOGAR','GASTO MENSUAL EN LINEA TELEFÓNICA', 'GASTO MENSUAL POR PAGO DE INTERNET' ]

Realizar la simulación de la carga de datos utilizando la siguiente información adicional:


1.   El 70 % de los encuestados son jóvenes entre 15 y 21 años, el 10% son mayores de 60 años. Sugerencia, utilizar esto para generar los DNI
2.   Los encuestados son de la provincia de Cordoba. Además el 80 % de la población de esa provincia tiene un grupo familiar de 5 personas. Con ingresos mensuales promedio de $258.000
2.   Cerca del 80 % de los encuestados Gastan el 90% de sus ingresos, el resto gasta el 80% de sus ingresos.Ayuda, para simular la cantidad de miembros utilizar la distribución normal.
3.  Del gasto Total por hogar se sabe que se abona un 15 % en telefonía y un 7.8% en servicio de internet.

Con la información proporcionada, realizar la simulación de las encuestas de 100 y 1000 personas.

*   Calcular el valor medio y desvio estandar de los gastos mensuales, ingresos mensuales y pagos de servicios.
*   Plotear los datos de los ingesos , y gastos en gráficos de barra por separado y evidenciar la media y desvio estandar.
*   Calcular la proporciona de jóvenes menores a 21 encuestados. Mostrar sus datos de grupos familiares, gastos en servicios de internet y telefonía. Concluir algo en base a esta información.
*   Realizar un histograma con el dato del grupo familiar, y concluir algo acerca de la distribución de probabilidad.




```python
import numpy as np
import matplotlib.pyplot as plt
from importlib import reload    #"solucion" problemas con plt.title (str object not callable)
plt=reload(plt)


#Primero generamos los datos necesarios

def generar_datos(N):
  #-------DNI-------
  # Proporciones proporcionadas
  prop_jovenes = 0.70
  prop_mayores = 0.10
  prop_otros = 1 - prop_jovenes - prop_mayores
  # Generamos números de DNI para jóvenes entre 15 y 21 años
  dni_jovenes = np.random.randint(45000000, 50000000, int(N * prop_jovenes))

  # Generamos números de DNI para mayores de 60 años
  dni_mayores = np.random.randint(10000000, 15000000, int(N * prop_mayores))

  # Generamos números de DNI para el grupo "otros" 21-60
  dni_otros = np.random.randint(15000000, 45000000, int(N * prop_otros))

  # Combinamos todos los números de DNI generados pra armar la lista de dni
  dni = np.concatenate([dni_jovenes, dni_mayores, dni_otros])

  # Mezclamos el orden de los DNI generados
  np.random.shuffle(dni)
  #----Cantidad de miembros ----
  # Generamos la cantidad de miembros por hogar utilizando distribución normal
  cantidad_miembros = np.round(np.random.normal(5, 1.5, N)).astype(int)
  # ---Los ingresos mensuales por hogar---
  ingresos_mensuales = np.random.normal(258.000, 40.000, N)
  # Calculamos las proporciones de gasto respecto a los ingresos

  #---Gastos mensuales---

  prop_gasto_90 = 0.80  # El 80% gasta el 90% de sus ingresos
  prop_gasto_80 = 1 - prop_gasto_90  # El resto gasta el 80% de sus ingresos /En otras palabras el 20% gasta el 80%

  # Generamos el porcentaje de gasto de cada encuestado

  porcentaje_gasto = np.random.choice([0.9, 0.8], size=N, p=[prop_gasto_90, prop_gasto_80])

  # Calculamos los gastos mensuales de línea telefónica e internet por hogar

  gasto_total_hogar = ingresos_mensuales * porcentaje_gasto

  #---GASTO MENSUAL  TELEFÓNICA---

  # 15% de gastos en línea telefónica
  gasto_linea_telefonica = gasto_total_hogar * 0.15


  #7.8% en servicio de internet

  gasto_de_internet=gasto_total_hogar * 0.078

  #Generamos el arreglo con los datos de cada usuario encuestado

  datos = np.column_stack((dni,gasto_total_hogar,ingresos_mensuales,cantidad_miembros,gasto_linea_telefonica,gasto_de_internet ))
  return datos

#Ingresamos un valor para la cantidad de personas a encuestar
num=1000

datos = generar_datos(num)

# Mostramos algunos ejemplos de los datos generados
for i in range(5):
    print(f"DNI: {datos[i,0]}, Miembros en el hogar: {datos[i,3]}, "
          f"Ingresos mensuales: ${datos[i,2]:.2f}, Gastos mensuales de hogar: ${datos[i,1]:.2f}, "
          f"Gasto mensual en línea telefónica: ${datos[i,4]:.2f}"
          f"Gasto mensual en internet:${datos[i,5]:.2f}")

#Calculamos valor medio de los  gastos mensuales, ingresos mensuales y pagos de servicios

media_gastos_mensuales=np.mean(datos[:,1])

media_ingresos_mensuales=np.mean(datos[:,2])

media_pagos_servicios=(np.mean(datos[:,4]))+(np.mean(datos[:,5]))


print("La media de los gastos mensuales es: ",media_gastos_mensuales)
print("La media de los ingresos mensuales es: ",media_ingresos_mensuales)
print("La media de los pagos por servicios es: ",media_pagos_servicios)

#Calculamos el desvio estandar de los  gastos mensuales, ingresos mensuales y pagos de servicios

desvio_gastos_mensuales=np.std(datos[:,1])

desvio_ingresos_mensuales=np.std(datos[:,2])

desvio_pagos_servicios=(np.std(datos[:,4]))+(np.std(datos[:,5]))


#Graficamos datos de los ingresos y los gastos evidenciando la media y el desvio estandar

plt.hist(datos[:,2],bins=10,color='blue')
plt.title("Ingresos mensuales")
plt.ylabel('Personas')
plt.xlabel('Ingresos (miles $)')

plt.axvline(media_ingresos_mensuales, color='red', linestyle='dashed', linewidth=2, label='Media ingresos')
plt.axvline(media_ingresos_mensuales+desvio_ingresos_mensuales, color='orange', linestyle='dashed', linewidth=2, label=' Media  + Desvio estándar')
plt.axvline(media_ingresos_mensuales-desvio_ingresos_mensuales, color='orange', linestyle='dashed', linewidth=2, label='Media - Desvio estándar')

plt.legend()
plt.grid(True)
plt.show()


plt.hist(datos[:,1],bins=10,color='darkblue')
plt.title("Gastos mensuales")
plt.xlabel("Gastos (miles $)")
plt.ylabel("Personas")
plt.axvline(media_gastos_mensuales, color='red', linestyle='dashed', linewidth=2, label='Media gastos')
plt.axvline(media_gastos_mensuales+desvio_gastos_mensuales, color='orange', linestyle='dashed', linewidth=2, label='Media + desvio estándar')
plt.axvline(media_gastos_mensuales-desvio_gastos_mensuales, color='orange', linestyle='dashed', linewidth=2, label='Media - desvio estándar')
plt.legend()
plt.grid(True)

plt.show()



#Calculo de la proporcion de jovenes menores a 21 años encuestados

dni=np.array(datos[:,0].astype(int))
condicion_dni=((datos[:,0] > 45000000) & (datos[:,0]< 50000000))
menores21=datos[condicion_dni]

print("tipo menores 21",type(menores21))

# Proporcion de jovenes menores a 21 encuestados

proporcion_menores21=len(menores21)/len(datos)

#Mostramos la proporcion de jovenes menores a 21 años

print(f"La proporción de jóvenes menores a 21 años encuestados es: {proporcion_menores21}")

#Mostramos algunos datos de menores a 21 años (grupo familiar, gastos en internet y telefonia)

for i in range(5):

  print(f"DNI: {menores21[i,0].astype(int)}, Miembros en el hogar: {menores21[i,3]}, "
  f"Ingresos mensuales: ${menores21[i,2]:.3f}, Gastos mensuales de hogar: ${menores21[i,1]:.3f}, "
  f"Gasto mensual en línea telefónica: ${menores21[i,4]:.3f},"
  f"Gasto mensual en internet:${menores21[i,5]:.3f}")

print("--------Gráficos respecto menores de 21 años-------")
print("")


#Histograma del grupo familiar  jovenes menores a 21 años encuestados

plt.hist(menores21[:,3].astype(int),bins=10,color="green")
plt.title=("Histograma de grupo familiar menores a 21 encuestados")
plt.xlabel("Grupo familiar (miembros)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()


#Histograma de gastos en internet de jovenes menores a 21 años encuestados


plt.hist(menores21[:,5],bins=20,color='orange',alpha=0.7)
plt.title=("Gastos de internet")
plt.ylabel("Frecuencia")
plt.xlabel("Gastos en internet(miles $)")

plt.grid(True)
plt.show()



#Histograma de gastos en telefonia de jovenes menores a 21 años encuestados


plt.hist(menores21[:,4],bins=20,color='skyblue',alpha=0.7)
plt.title=("Gastos de telefonia")
plt.ylabel("Frecuencia")
plt.xlabel("Gastos en telefonia(miles $)")

plt.grid(True)
plt.show()


#----Conclusiones sobre los datos relacionados a jovenes menores a 21 años:

#Debido a como se generan los datos sabemos que  paroximadamente el 70 por ciento de los
#encuestados es menor de 21 años,la mayoria de ellos pertenece
# a grupos familiar con un total de 5 integrantes, con ingresos promedios de $258000
#gastos mensuales oscilando entre el 90 y 80 por ciento de los ingresos
#correspondiendo un 15% del gasto  a telefonía y un 7.8% en servicio de internet



print("--------Gráfico grupo familiar respecto todos los encuestados-------")
print("")

#Grafica del histograma con el dato del grupo familiar


plt.hist(datos[:,3].astype(int),bins=10,color='darkorange')

#plt.title('Cantidad integrantes grupos familiar por familia')
plt.title=("Histograma del grupo familiar")
plt.ylabel("Familias")
plt.xlabel("Integrantes grupo familiar")
media_grupo_integrantes=np.mean(datos[:,3])

plt.axvline(media_grupo_integrantes, color='red', linestyle='dashed', linewidth=2, label='Media de integrantes')

plt.legend()
plt.grid(True)
plt.show()

#----Conclusiones de la grafica dato del grupo familiar:
#De la grafica podemos concluir que la distribucion es practicamente normal ( observando cierta asimetria hacia a la izquierda)
#El promedio de integrantes del grupo familiar en Córdoba es de 5 personasy el  numero de integrantes
#numero de integrantes del grupo familiar  mas frecuente  oscila entre 4 a 6 personas"
# resultados similares tambien se observan cambiando el numero de encuestados (100,1000,etc)
```

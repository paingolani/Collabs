**Diccionario de datos:**


Gender(Género): indica el g{enero del paciente.Valores: 0 - masculino, 1 - femenino

Hemoglobin(Hemoglobina): La hemoglobina es una proteína de los glóbulos rojos que transporta oxígeno a los órganos y tejidos del cuerpo y dióxido de carbono de los órganos y tejidos a los pulmones.Se mide en  g/dl (gramos por decilitro).

MCH (HCM): es la abreviatura de "hemoglobina corpuscular media". Es la cantidad media en cada uno de los glóbulos rojos de una proteína llamada hemoglobina, que transporta oxígeno por todo el cuerpo.Se mide en picogramos(pg)

MCHC (CHCM) son las siglas en inglés de concentración media de hemoglobina corpuscular (CHCM en castellano). Es una medida de la concentración media de hemoglobina en el interior de un solo glóbulo rojo.Se mide en mg/dl

MCV (VCM) : significa volumen corpuscular medio. Un análisis de sangre MCV mide el tamaño medio de sus glóbulos rojos. Se mide en Fl (femtolitros)

Result(Resultados):Si el paciente padece anemia o no. Valores: 0- no anémico, 1-anémico




**Conceptos sobre la anemia y valores normales- anormales**


La anemia es una afección que se desarrolla cuando la sangre produce una cantidad inferior a la normal de glóbulos rojos sanos. El cuerpo no obtiene suficiente cantidad de sangre rica en oxígeno. La falta de oxígeno puede producir cansancio o debilidad. También puede provocar dificultad para respirar, mareos, dolores de cabeza o latidos cardíacos irregulares.

El médico puede diagnosticarla revisando factores de riesgo, el historial médico, la alimentación. En el análisis físico algunas síntomas posibles de anemia son lengua pálida o uñas quebradizas .El profesional de la salud también puede realizar un examen físico para buscar síntomas de anemia, como lengua pálida o uñas quebradizas.

Otros métodos de diagnóstico son los análisis de sangre, pruebas de medula ósea  (verifican si la médula ósea está sana y si produce cantidades normales de glóbulos).Como complemento para descartar otras afecciones o complicaciones debido a la anemia se puede indicar: colonoscopia (busca sangrado u otros problemas como tumores en el colon.), endoscopia(busca sangrado en el esófago, el estómago y la primera parte del intestino delgado) , pruebas genéticas(buscan cambios en los genes que controlan cómo su cuerpo produce glóbulos rojos), pruebas de orina comprueban los riñones están funcionando correctamente y también  se puede saber si hay algún sangrado en el tracto urinario), pruebas de niveles de hierro en sangre y de absorción, etc.



**Análisis de sangre**


Este análisis mide muchas partes diferentes de la sangre, incluidos los glóbulos rojos, los glóbulos blancos y las plaquetas.

•	Los niveles de glóbulos rojos más altos o más bajos de lo normal podrían ser un signo de anemia. Los glóbulos rojos transportan oxígeno desde los pulmones al resto del cuerpo.

•	Los niveles de hemoglobina más altos o más bajos de lo normal pueden ser un signo de anemia. La hemoglobina es una proteína rica en hierro en los glóbulos rojos que transporta oxígeno.

•	Los niveles de hematocrito demasiado bajos pueden ser un signo de anemia. El hematocrito es una medida de cuánto espacio ocupan los glóbulos rojos en la sangre.

•	Los niveles del volumen corpuscular medio (MCV) más altos o más bajos de lo normal pueden ser un signo de anemia. El MCV es una medida del tamaño promedio de los glóbulos rojos.




**Valores normales en las pruebas de sangre:**


•	Hemoglobina (Hb):   Sus valores naturales son 13,5-17,5 g/dL en hombres y 12-15,5 g/dL en mujeres.


•	VCM (volumen corpuscular medio):  Lo  normal es que el valor este entre 78-100 fL. El valor para el adulto suele rondar en 90 fl.

•	HCM (hemoglobina corpuscular media): . Valores normales oscilarán entre 27-34 pg (picogramos).

•	CHCM (Concentración de hemoglobina corpuscular media): el rango nornal es de  34 ± 2 mg/dl




**Valores anormales en los análisis de sangre que podrían indicar anemia:**


Hemoglobina

Por debajo de 12 g/dL en el caso de las mujeres y 14 g/dL en los hombres.



HCM

Podemos considerar un nivel bajo de HCM en sangre cuando la concentración de hemoglobina por glóbulo rojo es inferior a los 27pg en adultos y 25pg en niños. Este problema se denomina Hipocromía. Estos valores bajos también podrían ser debidos a padecer de talasemia (afección hereditaria en la sangre).Valores mas altos de 33 pg en adultos y 35 pcg en niños pueden indicar anemia hipercrómica.

CHCM

(Concentración de hemoglobina corpuscular  media) en microesferocitosis (anemia hemolítica más frecuente en pediatría) familiar está aumentada por sobre el límite alto de lo normal (36 mg/dl) en el 50% de los casos. Asimismo, los pacientes con anemia de células de Sickle  también tienen eritrocitos (globulos rojos) con aumento en la CHCM, debido a la deshidratación celular. La CHCM disminuida con valores menores a 30 mg/dl se considera hipocromía (tipo de anemia debido a insuficiencia de hemoglobina en los glóbulos rojos)


VCM

Según el rango  de valores que puede tomar permite hace una clasificación además en diferentes tipos de anemia (microcríticas,normocríticas, macrocríticas)

-VCM < 80 Fl
Podemos tener anemias microciticas (causada por déficit de hierro o inhabilidad de utilizar el hierro) ,existe posibilidad de talasemias. Más frecuente en niños entre 1 y 3 años.

-VCM normal ( 80-100 Fl ) (anemias normocriticas) secundarias a una enfermedad crónica, hemorragias agudas, etc.

-VCM alto (> 100 fl) (anemias macrociticas) producidas  por déficit de vitamina b12, hipotiroidismo, alcoholismo, enfermedad del hígado, pre leucemias. etc

**Análisis de matriz de correlación**

En la matriz de correlacion podemos apreciar que no existe correlacion positiva ni negativa siguiera débiles entre las variables en estudio.


```python
#----- Proceso para analisis de componentes principales ---

#Estandarizamos el dataframe de anemia

scaler = StandardScaler()
data_scaled = scaler.fit_transform(anemia_ds2) #Estandarizamos

columns_anemia = ['Hemoglobin', 'MCH', 'MCHC', 'MCV']
df_scaled = pd.DataFrame(data_scaled, columns=columns_anemia)
df_scaled

#Realizamos analisis de componente principales (PCA)
# Aplicamos el Análisis de Componente Principal (PCA)

pca = PCA()
objeto = pca.fit(data_scaled)

# Desviación estándar de cada componente principal.
print(objeto.singular_values_)

# Varianza explicada por cada Componente principal

print(objeto.explained_variance_)

# Mostramos la cantidad de variables
print(np.sum(objeto.explained_variance_))


# Varianza acumulada de cada componente principal

varianza_acumulada = np.cumsum(objeto.explained_variance_ratio_)
print("Varianza Acumulada de cada Componente Principal:")
print(varianza_acumulada)

explained_variance=pca.explained_variance_ratio_

plt.figure(figsize=(6, 4))
plt.bar(range(4), explained_variance, alpha=0.5, align='center',label='individual explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()

#Efectuamos el gráfico de la Varianza Acumulada

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(varianza_acumulada) + 1), varianza_acumulada, marker='o')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Acumulada')
plt.title('Varianza Acumulada de Componentes Principales')
plt.xticks(range(1, len(varianza_acumulada) + 1))
plt.grid(True)
plt.show()


# Matriz de Componentes Principales

componentes_matrix = pd.DataFrame(data=objeto.components_, columns=anemia_ds2.columns)
print("Matriz de Componentes Principales:")
print(componentes_matrix)

# Visualización de la Matriz de Componentes

plt.figure(figsize=(12, 6))
sns.heatmap(componentes_matrix, annot=True, cmap='coolwarm')
plt.xlabel('Variables Originales')
plt.ylabel('Componentes Principales')
plt.title('Matriz de Componentes Principales')
plt.show()

#No se sigue el analisis (reddimenciones, eleccion de componentes principales , etc) porque no se puede reducir componentes en este dataset

```

Si la función de varianza acumulada en PCA es lineal (es el caso que presenta el dataset analizado), esto significa que cada componente principal contribuye en aproximadamente la misma proporción a la varianza total explicada. En este caso, las primeras componentes principales no tienen una ventaja significativamente mayor en términos de explicar la varianza que las últimas componentes. Esto puede indicar que los datos pueden estar distribuidos de una manera que no permite una reducción efectiva en la dimensión.En nuestro caso en particular si se eligiese 3 componentes solo tendriamos cubierto aproximadamente un 77 %  de la contribucion a la varianza total,siendo que deberia ser al menos el 85 % según se explico en clase , por lo tanto no resulta factible reducir las dimensiones.

En el análisis de componentes principales, el objetivo es encontrar un conjunto más pequeño de componentes que expliquen la mayor parte de la variabilidad en los datos. Si la función de varianza acumulada es lineal, podría indicar que no hay una separación clara de importancia entre las componentes, pero eso no excluye la posibilidad de que se pueda reducir dimensiones para propósitos de simplificación o visualización, aunque con un costo de perder algo de varianza.

En definitiva aplicar PCA (Análisis de Componentes Principales) para reducir componentes puede no ser muy beneficioso cuando no hay correlación entre las variables como se da en el caso del dataset analizado. PCA se utiliza para capturar la variabilidad en los datos al proyectarlos en un espacio de menor dimensión. Al no estar correlacionadas las variables, la mayoría de la variabilidad ya estaría contenida en las propias variables y no se ganaría mucho al reducir dimensiones. En este caso, se decide entonces no reducir la dimensionalidad y se han analizado las propias variables.

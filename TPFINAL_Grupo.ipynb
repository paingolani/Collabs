**Diccionario de datos:**


Gender(Gnero): indica el g{enero del paciente.Valores: 0 - masculino, 1 - femenino

Hemoglobin(Hemoglobina): La hemoglobina es una protena de los glbulos rojos que transporta oxgeno a los rganos y tejidos del cuerpo y dixido de carbono de los rganos y tejidos a los pulmones.Se mide en  g/dl (gramos por decilitro).

MCH (HCM): es la abreviatura de "hemoglobina corpuscular media". Es la cantidad media en cada uno de los glbulos rojos de una protena llamada hemoglobina, que transporta oxgeno por todo el cuerpo.Se mide en picogramos(pg)

MCHC (CHCM) son las siglas en ingls de concentracin media de hemoglobina corpuscular (CHCM en castellano). Es una medida de la concentracin media de hemoglobina en el interior de un solo glbulo rojo.Se mide en mg/dl

MCV (VCM) : significa volumen corpuscular medio. Un anlisis de sangre MCV mide el tamao medio de sus glbulos rojos. Se mide en Fl (femtolitros)

**Conceptos sobre la anemia y valores normales- anormales**


La anemia es una afeccin que se desarrolla cuando la sangre produce una cantidad inferior a la normal de glbulos rojos sanos. El cuerpo no obtiene suficiente cantidad de sangre rica en oxgeno. La falta de oxgeno puede producir cansancio o debilidad. Tambin puede provocar dificultad para respirar, mareos, dolores de cabeza o latidos cardacos irregulares.

El mdico puede diagnosticarla revisando factores de riesgo, el historial mdico, la alimentacin. En el anlisis fsico algunas sntomas posibles de anemia son lengua plida o uas quebradizas .El profesional de la salud tambin puede realizar un examen fsico para buscar sntomas de anemia, como lengua plida o uas quebradizas.

Otros mtodos de diagnstico son los anlisis de sangre, pruebas de medula sea  (verifican si la mdula sea est sana y si produce cantidades normales de glbulos).Como complemento para descartar otras afecciones o complicaciones debido a la anemia se puede indicar: colonoscopia (busca sangrado u otros problemas como tumores en el colon.), endoscopia(busca sangrado en el esfago, el estmago y la primera parte del intestino delgado) , pruebas genticas(buscan cambios en los genes que controlan cmo su cuerpo produce glbulos rojos), pruebas de orina comprueban los riones estn funcionando correctamente y tambin  se puede saber si hay algn sangrado en el tracto urinario), pruebas de niveles de hierro en sangre y de absorcin, etc.



**Anlisis de sangre**


Este anlisis mide muchas partes diferentes de la sangre, incluidos los glbulos rojos, los glbulos blancos y las plaquetas.

	Los niveles de glbulos rojos ms altos o ms bajos de lo normal podran ser un signo de anemia. Los glbulos rojos transportan oxgeno desde los pulmones al resto del cuerpo.

	Los niveles de hemoglobina ms altos o ms bajos de lo normal pueden ser un signo de anemia. La hemoglobina es una protena rica en hierro en los glbulos rojos que transporta oxgeno.

	Los niveles de hematocrito demasiado bajos pueden ser un signo de anemia. El hematocrito es una medida de cunto espacio ocupan los glbulos rojos en la sangre.

	Los niveles del volumen corpuscular medio (MCV) ms altos o ms bajos de lo normal pueden ser un signo de anemia. El MCV es una medida del tamao promedio de los glbulos rojos.

**Anlisis de matriz de correlacin**

En la matriz de correlacion podemos apreciar que no existe correlacion positiva ni negativa siguiera dbiles entre las variables en estudio.


```python
#----- Proceso para analisis de componentes principales ---

#Estandarizamos el dataframe de anemia

scaler = StandardScaler()
data_scaled = scaler.fit_transform(anemia_ds2) #Estandarizamos

columns_anemia = ['Hemoglobin', 'MCH', 'MCHC', 'MCV']
df_scaled = pd.DataFrame(data_scaled, columns=columns_anemia)
df_scaled

#Realizamos analisis de componente principales (PCA)
# Aplicamos el Anlisis de Componente Principal (PCA)

pca = PCA()
objeto = pca.fit(data_scaled)

# Desviacin estndar de cada componente principal.
print(objeto.singular_values_)

# Varianza explicada por cada Componente principal

print(objeto.explained_variance_)

# Mostramos la cantidad de variables
print(np.sum(objeto.explained_variance_))


# Varianza acumulada de cada componente principal

varianza_acumulada = np.cumsum(objeto.explained_variance_ratio_)
print("Varianza Acumulada de cada Componente Principal:")
print(varianza_acumulada)

explained_variance=pca.explained_variance_ratio_

plt.figure(figsize=(6, 4))
plt.bar(range(4), explained_variance, alpha=0.5, align='center',label='individual explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()

#Efectuamos el grfico de la Varianza Acumulada

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(varianza_acumulada) + 1), varianza_acumulada, marker='o')
plt.xlabel('Nmero de Componentes Principales')
plt.ylabel('Varianza Acumulada')
plt.title('Varianza Acumulada de Componentes Principales')
plt.xticks(range(1, len(varianza_acumulada) + 1))
plt.grid(True)
plt.show()


# Matriz de Componentes Principales

componentes_matrix = pd.DataFrame(data=objeto.components_, columns=anemia_ds2.columns)
print("Matriz de Componentes Principales:")
print(componentes_matrix)

# Visualizacin de la Matriz de Componentes

plt.figure(figsize=(12, 6))
sns.heatmap(componentes_matrix, annot=True, cmap='coolwarm')
plt.xlabel('Variables Originales')
plt.ylabel('Componentes Principales')
plt.title('Matriz de Componentes Principales')
plt.show()

#No se sigue el analisis (reddimenciones, eleccion de componentes principales , etc) porque no se puede reducir componentes en este dataset

```

Si la funcin de varianza acumulada en PCA es lineal (es el caso que presenta el dataset analizado), esto significa que cada componente principal contribuye en aproximadamente la misma proporcin a la varianza total explicada. En este caso, las primeras componentes principales no tienen una ventaja significativamente mayor en trminos de explicar la varianza que las ltimas componentes. Esto puede indicar que los datos pueden estar distribuidos de una manera que no permite una reduccin efectiva en la dimensin.En nuestro caso en particular si se eligiese 3 componentes solo tendriamos cubierto aproximadamente un 77 %  de la contribucion a la varianza total,siendo que deberia ser al menos el 85 % segn se explico en clase , por lo tanto no resulta factible reducir las dimensiones.

En el anlisis de componentes principales, el objetivo es encontrar un conjunto ms pequeo de componentes que expliquen la mayor parte de la variabilidad en los datos. Si la funcin de varianza acumulada es lineal, podra indicar que no hay una separacin clara de importancia entre las componentes, pero eso no excluye la posibilidad de que se pueda reducir dimensiones para propsitos de simplificacin o visualizacin, aunque con un costo de perder algo de varianza.

En definitiva aplicar PCA (Anlisis de Componentes Principales) para reducir componentes puede no ser muy beneficioso cuando no hay correlacin entre las variables como se da en el caso del dataset analizado. PCA se utiliza para capturar la variabilidad en los datos al proyectarlos en un espacio de menor dimensin. Al no estar correlacionadas las variables, la mayora de la variabilidad ya estara contenida en las propias variables y no se ganara mucho al reducir dimensiones. En este caso, se decide entonces no reducir la dimensionalidad y se han analizado las propias variables.
